## Abstract <a class="anchor" id="abstract"></a>
In this report, we undertake a comprehensive exploration of three-wave nonlinear mixing amidst longitudinal discrepancies in phase-matching conditions, following the study presented by Christophe Dorrer in the paper titled "Impact of Longitudinal Phase-Matching Variations on Three-Wave Nonlinear Interactions". Our methodology involves quantifying the performance of second-harmonic generation and optical parametric amplification through the application of normalized equations and a polynomial representation of wave-vector mismatch as a function of the longitudinal coordinate. These modeling outcomes are utilized to estimate the impact of spatial variations in wave-vector mismatch, which have been empirically observed for five partially deuterated potassium dihydrogen phosphate (DKDP) crystals. This study discerns that minor longitudinal irregularities in crystals with comparable quality are unlikely to significantly affect their suitability for second-harmonic generation and optical parametric amplification.

## 1. Introduction <a class="anchor" id="introduction"></a>

Three-wave mixing corresponds to the interaction of waves at frequencies $\omega_1$, $\omega_2$, and $\omega_3$ = $\omega_1$ + $\omega_2$. It is widely used for nonlinear frequency conversion, e.g., resulting in an output wave at frequency $2\omega$ generated from the input wave at frequency $\omega$ in second-harmonic generation (SHG) (which is usually implemented when the source is not very efficient at higher wavelengths or lower frequencies), and parametric amplification (OPA) resulting in energy transfer from the input pump at frequency $\omega_3$ to the signal and idler at frequencies $\omega_1$ and $\omega_2$. <br>
<br>
Efficient three-wave mixing requires phase matching. We saw this in couple of examples in class - most prominently in SHG and SFG in undepleted pump approximation. This requirement is generally expressed as $\Delta \bf{k} = \bf{k_3} – \bf{k_2} – \bf{k_1} ≈ 0$, where $\bf{k_j}$ is the wave vector at frequency $\omega_j$. For collinear interactions (which is what we discussed in class), this relation can be projected on the common propagation axis, which is taken to be z by convention, but geometric angle-dependent terms introduced by noncollinear interactions lead to additional flexibility. In most cases, the resulting $\Delta \bf{k}$ does not depend on the transverse coordinates (in the plane perpendicular to the propagation direction) or longitudinal coordinate (along the propagation direction), based on the assumption that the local properties of the nonlinear medium are spatially uniform. Therefore, generally, $\Delta \bf{k}$ is independent of all coordinates $x,y,z$. However, in the cases where we can control the magnitude of the wave vector for individual waves, longitudinal variations in wave-vector mismatch can enhance the nonlinear conversion, i.e., increase the conversion efficiency.

### A note on the choice of KDP/DKDP crystals for this paper
Most nonlinear-optics applications are based on crystals with dimensions of the order of 1 cm, but much larger crystals are required for high-energy laser pulses. Frequency conversion of high-energy laser systems is based on large-aperture potassium dihydrogen phosphate (KDP) and its partially deuterated isomorph, DKDP. DKDP can also support broadband optical parametric amplification (OPA) of a signal around 910 nm using a pump pulse generated by frequency doubling of an Nd:doped laser. High-energy amplification stages rely on crystals with thickness and/or aperture larger than 5 cm, and much larger crystals would be required for the forecasted scaling to tens of petawatts. KDP and its isomorphs are currently the easiest crystals to grow at very large aperture, spatially uniform even with a large apperture, and possess local variations in deuteration level, with all of these properties analyzed in previous works.

## 2. Simulation Framework

### 2.1 Longitudinal variations in phase-matching conditions

In our analysis, we explore the phenomenon of non-linear mixing involving three waves characterized by frequencies $\omega_1$, $\omega_2$, and $\omega_3 = \omega_1 + \omega_2$. We start by making the assumption that the walk-off effect, which pertains to the divergence in the Poynting vectors of these waves, can be considered negligible. Furthermore, we assume that the optical index, denoted as $n_j$, associated with each of the waves at the respective frequencies $\omega_j$, remains invariant with respect to the longitudinal coordinate '$z$'.

As these waves propagate from the initial reference plane at '$z = 0$' to a plane situated at coordinate '$z$', they accumulate phase shifts characterized by the term $k_jz$, where the propagation constant, denoted as $k_j$, is defined as $k_j = \frac{n_j \omega_j}{c}$. In the presence of longitudinal variations in the optical index, we derive a similar set of equations, under the condition that these variations in $n_j$ remain small within the region where $z \approx \lambda_j$, where $\lambda_j$ represents the wavelength associated with each of the waves.

In this case, the accumulation in the phase is given by
\begin{equation}
\varphi_i(z) = \int_0^z k_j(z') dz'
\end{equation}
The phase mismatch at coordinate $z$ is therefore $\Delta \varphi(z) = \varphi_3(z) - \varphi_2(z) - \varphi_1(z)$. Note the convention for the definition of $\Delta \varphi(z)$. The propagation equations then come out to be


\begin{equation}
\frac{\partial E_1}{\partial z} = i \frac{\omega_1 d_{\text{eff}}}{c n_1} E_3 E_2^* \exp[i\Delta\varphi(z)]
\end{equation}

\begin{equation}
\frac{\partial E_2}{\partial z} = i \frac{\omega_2 d_{\text{eff}}}{c n_2} E_3 E_1^* \exp[i\Delta\varphi(z)]
\end{equation}

\begin{equation}
\frac{\partial E_3}{\partial z} = i \frac{\omega_3 d_{\text{eff}}}{c n_3} E_1 E_2 \exp[-i\Delta\varphi(z)]
\end{equation}

here, where $d_{eff}$ is the effective nonlinearity and $c$ is the speed of light.

It is important to highlight that a factor of 2 on the right-hand side of the equations is absent. This discrepancy arises from the author's convention, wherein the electric field is expressed as $E_j(z,t) = \text{Re}(E_j e^{i(\phi_j(z) - \omega_j t)})$, as opposed to the convention we've used in class, which employs the expression $E_j(z,t) = 2 \cdot \text{Re}(E_j e^{i(\phi_j(z) - \omega_j t)})$ or equivalently $E(z,t) = E_j e^{i(\phi_j(z) - \omega_j t)}$ $+$ $c.c.$

\begin{equation}
\Delta k(x,y,z) =[\Delta k(x,y,z) - \langle\Delta k(x,y,z)\rangle] + \langle\Delta k(x,y,z)\rangle
\end{equation}

Here, $\langle\Delta k(x, y, z)\rangle$ represents the average of $\Delta k(x, y, z)$ calculated over the entire crystal length denoted as 'L'. The bracketed term showcases the longitudinal variation in wave-vector mismatch concerning its deviation from the average value at these particular coordinates (x, y).

In the subsequent sections, our attention is directed towards quantifying the influence exerted by longitudinal variations in wave-vector mismatch on a nonlinear process. This process can be categorized as either being nominally phase matched, where $\langle\Delta k(x,y,z)\rangle$ equals zero, or as detuned from the phase matching condition, denoted by $\langle\Delta k(x,y,z)\rangle$ not equal to zero.

### 2.2 Normalized Equations
As was demonstrated during our classroom discussions, for the sake of simplifying the calculations, we introduce the concept of normalized fields, denoted as $e_j$. These normalized fields are defined as:

\begin{equation}
e_j = E_j{\sqrt{\frac{\varepsilon_0 c n_j}{2\hbar \omega_j}}} \kappa
\end{equation}

Here, $e_j$ represents the normalized field, and $\kappa$ stands for a constant. The concept of field intensity is concurrently introduced as:

\begin{equation}
I_j = \frac{\varepsilon_0 c n_j}{2} |E_j|^2
\end{equation}

It is again important to note that the factor of 2 in the denominator is introduced in accordance with the chosen convention for the electric field, as elaborated in section 2.1. The selected form of normalization is designed such that $|e_j|^2$ is directly proportional to the number of photons at their respective frequencies.

Additionally, we normalize the coordinate '$z$' by expressing it as '$z = LZ$', where '$L$' represents the length of the nonlinear medium, and '$Z$' is confined within the interval [0, 1]. Consequently, the normalized phase mismatch as a function of '$Z$' can be expressed as:
    
 \begin{equation}
 \Delta \Phi (Z) = \Delta \varphi(LZ)
 \end{equation}

We employ this to scale wave-vector variations ($\Delta k$) determined from one crystal to normalized wave vector variations($\Delta K$) as,
\begin{equation}
\Delta K (Z) = L\Delta k(LZ)
\end{equation}

To determine the constant $\kappa$, we substitute the normalized field amplitudes into the coupled field equations. Consequently, the first equation becomes:

\begin{equation}
\frac{\partial e_1}{\partial Z} = i\frac{Ld_{eff}}{\kappa} \sqrt{\frac{2\hbar\omega_1\omega_2\omega_3}{n_1n_2n_3c^3\epsilon_0}} e_3 e_2^* \exp[i\Delta\Phi(z)]
\end{equation}

By setting $\kappa$ equal to $Ld_{eff}\sqrt{\frac{2\hbar\omega_1\omega_2\omega_3}{n_1n_2n_3c^3\epsilon_0}}$, we obtain three equations as follows:

\begin{equation}
\frac{\partial e_1}{\partial Z} = i e_3 e_2^* \exp[i\Delta\Phi(z)]
\end{equation}

\begin{equation}
\frac{\partial e_2}{\partial Z} = i e_3 e_1^* \exp[i\Delta\Phi(z)]
\end{equation}

\begin{equation}
\frac{\partial e_3}{\partial Z} = i e_1 e_2 \exp[-i\Delta\Phi(z)]
\end{equation}

These equations are versatile and can be applied to describe the nonlinear mixing of three nondegenerate waves in various contexts. For instance, they are suitable for modeling the evolution of the signal, idler, and pump fields in an Optical Parametric Amplifier (OPA) without the necessity to account for the individual wavelengths and corresponding optical indices.

Additionally, this same system of equations can be employed in the context of Type II Second Harmonic Generation (SHG), with $e_1$ and $e_2$ representing orthogonally polarized waves at frequency $\omega$, and $e_3$ representing the field at the upconverted frequency $2\omega$.

For Type I Second Harmonic Generation (SHG), these equations can be adapted to account for the degeneracy of two fields at the fundamental frequency. In this case, where $e_1$ represents the field at the fundamental frequency $\omega_1$ and $e_3$ represents the field at the upconverted frequency $2\omega$, the equations are as follows:

\begin{equation}
\frac{\partial e_1}{\partial Z} = i e_3 e_1^* \exp[i\Delta\Phi(z)]
\end{equation}

\begin{equation}
\frac{\partial e_3}{\partial Z} = \frac{i}{2} e_1^2 \exp[-i\Delta\Phi(z)]
\end{equation}


### 2.3 SHG and OPA in ideal phase-matching conditions
In this section, we utilize the previously derived normalized coupled field equations, outlined in Section 2.2, to generate plots illustrating the photon conversion efficiency for both phase matched Second Harmonic Generation (SHG) and Optical Parametric Amplification (OPA). The photon conversion efficiency is quantified as the ratio of the number of photons in the output wave (which corresponds to $2\omega$ in the case of SHG, and $\omega_1$ for OPA) to the number of photons in the input wave (i.e., $\omega$ for SHG, and $\omega_3$ for OPA).

Given that our normalized fields have been defined in a manner such that their squared modulus is directly proportional to the number of photons at their respective frequencies, we denote $i_j$ as $|e_j|^2$.

### 2.3.1 SHG
The photon conversion efficiency is mathematically described as the ratio of the output intensity at $2\omega$, denoted as $i_3(1)$, to the input intensity at $\omega$, represented as $i_1(0)$. Here, the lowercase $i$ signifies the squared modulus of the normalized field amplitudes $e_i$.

To numerically solve the coupled differential equations, we have employed the fourth-order Runge-Kutta method.

```python
def de1dz(Z,e1,e3):
  return 1j * e3 * np.conj(e1)
def de3dz(Z,e1,e3):
  return 0.5j * e1**2
```

```python
def RungeKuttaCoupled(Z, e1, e3, dZ, de1dZ, de3dZ):

    k1 = dz*de1dz(Z,e1,e3)
    h1 = dz*de3dz(Z,e1,e3)
    k2 = dz*de1dz(Z+dz/2., e1+k1/2., e3+h1/2.)
    h2 = dz*de3dz(Z+dz/2., e1+k1/2., e3+h1/2.)
    k3 = dz*de1dz(Z+dz/2., e1+k2/2., e3+h2/2.)
    h3 = dz*de3dz(Z+dz/2., e1+k2/2., e3+h2/2.)
    k4 = dz*de1dz(Z+dz, e1+k3, e3+h3)
    h4 = dz*de3dz(Z+dz, e1+k3, e3+h3)

    e1 = e1 + 1./6.*(k1+2*k2+2*k3+k4)
    e3 = e3 + 1./6.*(h1+2*h2+2*h3+h4)
    Z = Z + dz

    return Z, e1, e3
```

```python
e31e10 = []
for i in range(len(e10)):
    e30 = 0+0j
    e10_a = e10[i]
    dz = 0.01
    Z0 = 0
    Z_end = 1
    e1 = e10_a
    e3 = e30
    Z = Z0
    Z_list = [Z0]
    e1_list = [e10_a]
    e3_list = [e30]
    while Z <= Z_end:
      Z, e1, e3 = RungeKuttaCoupled(Z, e1, e3, dz, de1dz, de3dz)

      Z_list.append(Z)
      e1_list.append(e1)
      e3_list.append(e3)
    e31 = e3_list[-1]
    e31e10.append(e31*np.conj(e31) / np.abs(e10[i])**2)
```

## 3. Simulations of three-wave nonlinear mixing in the presence of longitudinal variations in wave-vector mismatch
For simulations, we express the wave-vector mismatch ∆K as a sum of Legendre polynomials Pj having their argument between –1 and 1 and coefficients $a_j$ that are randomly distributed in a
specific range [$–\rho, \rho$], such that

<center> $\Delta K(Z) = \sum\limits_{j=1}^N a_jP_j(2Z − 1)$</center>

(Largest coefficient range corresponds to $\rho$ =1). We consider polynomials with maximal orders equal to 1, 2, 3 and 5 to assess the sensitivity to the
frequency of the longitudinal spatial variations (appearance of longitudinal variations over z). This can represent the bracketed term in the expansion of $ \Delta k(x,y,z)$ as given in section 2.1, because the integral of these Legendre polynomials over [-1,1] is equal to 0 and so is the integral of the bracketed term. Non-phase-matched conditions correspond to the sum of $\Delta K(Z)$ and a constant $\Delta\Phi_0$ that does not depend on Z. Integration of this constant between 0 and 1 gives phase mismatch $\Delta\Phi_0$ in the absence of longitudinal wave-vector mismatch variations.

###3.2 Second-harmonic generation
In the following code, we plot the relative efficieny for an SHG process at $\Delta\Phi_0 = 0$ and $i_1(0)$ = 3, resulting in an intensity conversion efficiency  of 0.35 (70%), as a function of the range of legendre polynomial coefficients $\rho$. The relative efficiency is defined as the ratio of photon conversion efficiency in the presence of longitudinal variations to the nominal photon conversion efficiency (i.e., in the case of perfect phase matching). We run the simulation 100 times, where each iteration corresponds to a different set of legendre coefficients and therefore, a different $\Delta K (Z)$ corresponding to each $\rho$.  We observe a quadratic trend in the variation of the relative efficiency as a function of $\rho$ as mentioned in the paper. For each N value, we take 5 values of $\rho$ and find the corresponding spread in relative efficiency after obtaining the best quadratic fit.

```python
def delta_K(Z,N,a_arr):
  s = 0
  if N != 0:
    for i in range(0,N):
      term = a_arr[i]*eval_legendre(i+1,2*Z-1)
      s = s + term
  else:
    s = a_arr[0]
  #print(a_arr,s)
  return s
```

```python
def delta_phi(Z,N,a_arr):
  return quad(delta_K,0,Z,args = (N,a_arr))[0]
```

```python
def de1dz(Z,e1,e3,N,a_arr):
  return 1j * e3 * np.conj(e1)*np.e**(1j*delta_phi(Z,N,a_arr))
def de3dz(Z,e1,e3,N,a_arr):
  return 0.5j * e1**2*np.e**(-1j*delta_phi(Z,N,a_arr))
```

```python
def delta_phi(Z,N,a_arr):
  return quad(delta_K,0,Z,args = (N,a_arr))[0]
```

```python
def de1dz(Z,e1,e2,e3,N,a_arr):
  return 1j * e3 * np.conj(e2)*np.e**(1j*delta_phi(Z,N,a_arr))
def de2dz(Z,e1,e2,e3,N,a_arr):
  return 1j * e3 * np.conj(e1)*np.e**(1j*delta_phi(Z,N,a_arr))
def de3dz(Z,e1,e2,e3,N,a_arr):
  return 1j * e1 * e2*np.e**(-1j*delta_phi(Z,N,a_arr))
```

```python
def RungeKuttaCoupled(Z, e1, e2, e3, dZ, de1dZ, de2dZ, de3dZ, N,a_arr):

    k1 = dz*de1dz(Z,e1,e2,e3,N,a_arr)
    l1 = dz*de2dz(Z,e1,e2,e3,N,a_arr)
    h1 = dz*de3dz(Z,e1,e2,e3,N,a_arr)
    k2 = dz*de1dz(Z+dz/2., e1+k1/2., e2+l1/2., e3+h1/2.,N,a_arr)
    l2 = dz*de2dz(Z+dz/2., e1+k1/2., e2+l1/2., e3+h1/2.,N,a_arr)
    h2 = dz*de3dz(Z+dz/2., e1+k1/2., e2+l1/2., e3+h1/2.,N,a_arr)
    k3 = dz*de1dz(Z+dz/2., e1+k2/2., e2+l2/2., e3+h2/2.,N,a_arr)
    l3 = dz*de2dz(Z+dz/2., e1+k2/2., e2+l2/2., e3+h2/2.,N,a_arr)
    h3 = dz*de3dz(Z+dz/2., e1+k2/2., e2+l2/2., e3+h2/2.,N,a_arr)
    k4 = dz*de1dz(Z+dz, e1+k3, e2+l3, e3+h3,N,a_arr)
    l4 = dz*de2dz(Z+dz, e1+k3, e2+l3, e3+h3,N,a_arr)
    h4 = dz*de3dz(Z+dz, e1+k3, e2+l3, e3+h3,N,a_arr)

    e1 = e1 + 1./6.*(k1+2*k2+2*k3+k4)
    e2 = e2 + 1./6.*(l1+2*l2+2*l3+l4)
    e3 = e3 + 1./6.*(h1+2*h2+2*h3+h4)
    Z = Z + dz

    return Z, e1, e2, e3
dz = 0.01
```

We now analyze the variation of efficiency with the root mean square of the wave-vector mismatch $\Delta K$ denoted by $\sigma$.

```python
def DelKsq(Z,N,a_arr):
  s = 0
  if N != 0:
    for i in range(0,N):
      term = a_arr[i]*eval_legendre(i+1,2*Z-1)
      s = s + term
  else:
    s = a_arr[0]
  #print(a_arr,s)
  return s*s
```

```python
def DelK_RMS(N,a_arr):
  return np.sqrt(quad(DelKsq,0,1,args=(N,a_arr))[0])
```

```python
def delta_K_phi_0(Z,N,a_arr,phi_0):
  #print(a_arr)
  s = 0
  if N != 0:
    for i in range(0,N):
      term = a_arr[i]*eval_legendre(i+1,2*Z-1)
      s = s + term
  else:
    s = 0
  #print(a_arr,s)
  return s + phi_0
```

```python
def delta_phi_phi_0(Z,N,a_arr,phi_0):
  #print(a_arr)
  return quad(delta_K_phi_0,0,Z,args = (N,a_arr,phi_0))[0]
```

```python
def de1dz(Z,e1,e2,e3,N,a_arr,phi_0):
  return 1j * e3 * np.conj(e2)*np.e**(1j*delta_phi_phi_0(Z,N,a_arr,phi_0))
def de2dz(Z,e1,e2,e3,N,a_arr, phi_0):
  return 1j * e3 * np.conj(e1)*np.e**(1j*delta_phi_phi_0(Z,N,a_arr,phi_0))
def de3dz(Z,e1,e2,e3,N,a_arr, phi_0):
  return 1j * e1 * e2*np.e**(-1j*delta_phi_phi_0(Z,N,a_arr,phi_0))
```

```python
def RungeKuttaCoupled_phi0(Z, e1, e2, e3, dZ, de1dZ, de2dZ, de3dZ, N,a_arr, phi_0):
  k1 = dz*de1dz(Z,e1,e2,e3,N,a_arr,phi_0)
  l1 = dz*de2dz(Z,e1,e2,e3,N,a_arr,phi_0)
  h1 = dz*de3dz(Z,e1,e2,e3,N,a_arr,phi_0)
  k2 = dz*de1dz(Z+dz/2., e1+k1/2., e2+l1/2., e3+h1/2.,N,a_arr,phi_0)
  l2 = dz*de2dz(Z+dz/2., e1+k1/2., e2+l1/2., e3+h1/2.,N,a_arr,phi_0)
  h2 = dz*de3dz(Z+dz/2., e1+k1/2., e2+l1/2., e3+h1/2.,N,a_arr,phi_0)
  k3 = dz*de1dz(Z+dz/2., e1+k2/2., e2+l2/2., e3+h2/2.,N,a_arr,phi_0)
  l3 = dz*de2dz(Z+dz/2., e1+k2/2., e2+l2/2., e3+h2/2.,N,a_arr,phi_0)
  h3 = dz*de3dz(Z+dz/2., e1+k2/2., e2+l2/2., e3+h2/2.,N,a_arr,phi_0)
  k4 = dz*de1dz(Z+dz, e1+k3, e2+l3, e3+h3,N,a_arr,phi_0)
  l4 = dz*de2dz(Z+dz, e1+k3, e2+l3, e3+h3,N,a_arr,phi_0)
  h4 = dz*de3dz(Z+dz, e1+k3, e2+l3, e3+h3,N,a_arr,phi_0)

  e1 = e1 + 1./6.*(k1+2*k2+2*k3+k4)
  e2 = e2 + 1./6.*(l1+2*l2+2*l3+l4)
  e3 = e3 + 1./6.*(h1+2*h2+2*h3+h4)
  Z = Z + dz

  return Z, e1, e2, e3
```

```python
def plot_phi0(N,I,k):
  del_phi_0 = np.linspace(0,4*np.pi,500)
  my_list = []
  #N = 5
  for i in del_phi_0:
    if N!=0:
      a_arr = np.random.uniform(-0.5,0.5,N)
    else:
      a_arr = [i]
    e10 = 10**(-1.5)+0j
    e20 = 0+0j
    e30 = np.sqrt(I)+0j
    dz = 0.01
    Z0 = 0
    Z_end = 1
    e1 = e10
    e2 = e20
    e3 = e30
    Z = Z0
    Z_list = [Z0]
    e1_list = [e10]
    e2_list = [e20]
    e3_list = [e30]
    while Z <= Z_end:

      Z, e1, e2, e3 = RungeKuttaCoupled_phi0(Z, e1, e2, e3, dz, de1dz, de2dz, de3dz, N,a_arr, i)

      Z_list.append(Z)
      e1_list.append(e1)
      e2_list.append(e2)
      e3_list.append(e3)
  #print(e3_list)
    if I == 45.5:
      my_list.append(e1_list[-1]*np.conj(e1_list[-1])/(45.5))
    else:
      my_list.append(e1_list[-1]*np.conj(e1_list[-1])/(32.5*0.5))
  if k == 0:
    plt.scatter(del_phi_0, my_list,c = col[N],s=0.1,label = "N = {}".format(N))
  else:
    plt.scatter(del_phi_0, my_list,c = col[N],s=0.1)
  #for i in range(len(rho)):
    #plt.plot(rho[i],my_list[i],"ro")
  #m = np.linspace(0,1,100)
  #popt, pcov = curve_fit(f, del_phi_0, my_list)
  #plt.plot(m, f(m, *popt))
```

To plot the average magnitude of the wave-vector mismatch difference at all transverse points within the aperture relative to its value at the center for crystals 1 to 5, the following steps are employed:

1. Calculate the absolute difference between $\Delta k_{SHG}(x, y)$ and the reference value $\Delta k_{SHG}(0,0)$ for every individual point within the specified aperture. This quantifies how much each point's wave-vector mismatch at every spatial coordinate deviates from the central value within the defined area.
2. Organize these differences into 'bins' based on the distance of each point from the center of the aperture. This distance is determined by the spatial position of each point relative to the central point (0,0).
3. Compute the average of the absolute differences within each bin. This reveals the typical change in measured wave-vector mismatch for points located at similar distances from the center.
4. Present the results graphically, to illustrate how the average wave-vector mismatch varies as you move away from the central point within the aperture. This graphical representation helps visualize the spatial distribution of wave-vector mismatch.

**Note**: Multiple scans were performed for the crystals, for example, with the input face being changed (either S1 or S2), to ensure the consistency and reliability of the findings, although these measurements may yield slightly different results due to variations introduced by the surface features on the input face of the crystal, which can impact the phase matching within the crystal, whereas they have no impact when the same face is the output face. ### 4.2 Second-harmonic generation at 1053 nm

Integration length $L_{SHG}=12mm$ (Typical SHG crystals for large-scale high-energy systems are $\sim 12mm$ long.)

**lineouts**: Representative of the variations in $\Delta k$ over that chosen length; there are infinite number of ways 1D lineouts could be extracted from the provided 2D maps, so it is an arbitrary choice for data that is typical of what is observed.

For the small crystals, four lineouts are used; two horizontal and two vertical, aligned along x=0 or y=0. For the large crystals, eight lineouts are chosen; four horizontal and four vertical, aligned along x=0 or y=0.

Hence, total number of sets $= 3*4 + 2*8 = 28$ sets of $\Delta k$ for SHG




Fitting of each of the generated lineouts by a sum of Legendre polynomial
$$\Delta K(Z) = \sum_{j=1}^{N} b_{j} P_{j} (2Z-1) $$

**Note**: Utility of plotting fitted coefficients with polynomial degrees is to show that the coefficients have low values, and in particular that the observed variations are mostly low order.

### 4.4 Wave-vector mismatch at other wavelengths

**Case1**: SHG

The experimentally determined wave-vector mismatch variations at 1053 nm using SHG $(\lambda_{3} = \lambda_1/2)$ serve as a basis for evaluating the wave-vector mismatch at different wavelengths by considering the influence of variations in the internal angle $(\delta θ)$ and deuteration level $(\delta X)$.

In partially deuterated KDP crystal, the phase-mismatch variation can be induced by variations in either the local deuteration level or the internal crystal angle, or any combination of these variations,

$\Delta k_{SHG} = \frac{2π}{λ_3}n_e(λ_3, θ + δθ, X + \delta X) - \frac{4π}{λ_1} n_o(λ_1, X + \delta X)]$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$=\frac{4π}{λ_1}[n_e(λ_3, θ + δθ, X + \delta X) - n_o(λ_1, X + \delta X)]$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$= \frac{4π}{λ_1}[∂_{θ, 3}n_eδθ + (∂_{X, 3}n_e  - ∂_{X, 1}n_o)δX]  \; \; \; \; \; \; \; \; \; \; ---(a)$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$= ∂_θ\Delta k_{SHG}δθ + ∂_X\Delta k_{SHG}δX \; \; \; \; \; \; \; \; \; \; \; \; \; \; \; \; \;\; \; \;\;\; \; ---(b)$

where, where the variations in the local angle of the crystal axis relative to the propagation axis $\delta \theta$ and the local deuteration level $\delta X$ depend on the longitudinal position

$∂_{θ, 3}n_e =$ partial derivative wrt the angle $θ$ of the extraordinary index at the wavelength $λ_3$, and

$∂_{X,j}n =$ partial derivative wrt the deuteration level $X$ of the index $n$ (either ordinary or extraordinary) at the wavelength $λ_j$.

Comparing equations (a) and (b),  <br>
$$∂_θ\Delta k_{SHG} = \frac{4π}{λ_1} (∂_{θ, 3}n_e), \; \; \; \; \; \; \; \; \; \; \; \; \; \; \; \; \; \; \; \;---(c)$$

The partial derivative of wave-vector mismatch relative to $\theta$, describes the sensitivity of the wave-vector mismatch with respect to an internal change in crystal-axis angle.

To estimate the above partial derivative wrt phase-matching angle, extraordinary index must be generalized for arbitrary $\theta$. Hence, extraordinary index at an angle $\theta$:

$$n_e(\lambda_3, \theta, X) = \frac{1}{\sqrt{\left (\frac{\cos{\theta}}{n_o(\lambda_3, X)}\right )^{2}+ \left (\frac{\sin{\theta}}{n_e(\lambda_3, X)}\right )^{2}}}$$

### Sellmeier Equations
For nonlinear optical applications like SHG and THG, the refractive indices $n_o$ and $n_e$ are calculated from the Sellmeier equations

$$n^{2} = 1 + \frac{a_{1}}{\lambda_{3}^{2}-b_{1}^2} + \frac{a_{2}\lambda_{3}^{2}}{\lambda_{3}^{2}-b_{2}^{2}}$$

And, the modified version of the equation by Zernike (Modified Sellmeier equation),
$$n^{2} = A_{0} + \frac{A_{1}}{\lambda_{3}^{2}-B_{1}^2} + \frac{A_{2}\lambda_{3}^{2}}{\lambda_{3}^{2}-B_{2}^{2}}$$

The second term of this equation represents the infrared contribution and the last term represents the adsorption contribution.

|$A_1$ | 0.010211 | 0.0087778|
|$B_1$| 0.011517 | 0.010903 |
| $A_2$ | 0.75554| 0.35343 |
|$B_2$ | 37.437 | 57.511 |

For SHG at 1053 nm in DKDP, $∂_{\theta, 3} n_{e}$ does not significantly depend on the deuteration level X; it is equal to –0.039 for X = 70% Ref.[2], which gives the final equation for (c) as

$$∂_θ\Delta k_{SHG} = \frac{4π}{λ_1} (-0.039)$$

```python
# Fig.8(a)
import numpy as np
import matplotlib.pyplot as plt

# Define the wavelength range (λ_1) from 1000 to 2000
wavelength_range = np.linspace(1000*1e-9, 1090*1e-9, 1000)

# Calculate ∂_θ Δk_SHG for each wavelength
d_angle = (4 * np.pi / wavelength_range) * (-0.039)

plt.figure(figsize=(8, 6))
plt.plot(wavelength_range, d_angle, label='∂_θΔk_SHG', color='blue')
plt.xlabel('$λ_1(nm)$')
plt.ylabel('$∂_θ Δk_{SHG}$')
plt.title('$∂_θ Δk_{SHG}$ vs $λ_1$')
#plt.grid()
plt.legend()

highlight_wavelength = 1053*1e-9
highlight_index = np.abs(wavelength_range - highlight_wavelength).argmin()
plt.scatter(wavelength_range[highlight_index], d_angle[highlight_index], color='black', s=100, label=f'λ_1 = {highlight_wavelength}')
plt.show()
```

Comparing equations (a) and (b),  <br>

$$∂_X\Delta k_{SHG} = \frac{4π}{λ_1} (∂_{X, 3}n_e  - ∂_{X, 1}n_o)\; \; \; \; \; \; \; \; \; \; \; \; \; \;---(d)$$

To estimate the above partial derivatives, refractive index dispersion must be generalized for arbitrary deuteration fractions

**Method1**: Requires either the interpolation or extrapolation of indices determined at reference deuteration levels. The most commonly reported approach is to calculate the susceptibility at a deuteration level x as a weighted average of the susceptibility at two reference levels $x_1$ and $x_2$:

$$n_j^2(\lambda, x) = \frac{n_j^2(\lambda, x_2)(x - x_2) + n_j^2(\lambda, x_1)(x_2 -x)}{x_2 - x_1}$$

where j stands for either o(ordinary axis) or e(extraordinary axis, in which case the index is also a function of the phase-matching angle $\theta$)

This approach, as mentioned in Ref.[13], although consistent with experimental data, lacks the fundamental confirmation obtained from precise experimental index measurements.

**Method2**: An alternative approach involves treating partially deuterated KDP as a mixture of pure KDP (x = 0%) and fully deuterated KDP (x = 100%), factoring in their mass density through the Lorentz–Lorenz equation, expressed as Ref.[3] (refractive index n of pDKDP with deuteration fraction x)

$$ \frac{n(x)^2 - 1}{n(x)^2 + 2} = \left( 1 - \frac{x - L}{H - L} \right) \frac{\rho(x)}{\rho(L)} \frac{n(L)^2 - 1}{n(L)^2 + 2} + \frac{x - L}{H - L} \frac{\rho(x)}{\rho(H)} \frac{n(H)^2 - 1}{n(H)^2 + 2}, \; \; \; \; \; \; \; (L \leq x \leq H)$$

where $n(L)$ and $\rho(L)$ are the absolute refractive index value and the mass density with a deuteration fraction of L, respectively. $n(H)$ and $\rho(H)$ follow similar definitions.

Simple interpolation to estimate the mass density $\rho(x)$ for a deuteration fraction $x$
$$\rho(x) = (1-x)\rho_{KDP} + x \rho_{DKDP}$$

where $\rho_{KDP} = \rho(0) = 2.332 g/cm^{3}$ (pure KDP) and $\rho_{DKDP} = \rho_{1} = 2.355 g/cm^3$ (pure DKDP)

Remarkably, this approach produces results akin to the susceptibility average method, likely owing to the similar mass densities of KDP and DKDP.

hence, the bracketed quantity in eq(d) is equal to $-3×10^{−3}$ in 70% DKDP (Ref.[2]), which gives

$$∂_X\Delta k_{SHG} = \frac{4π}{λ_1} (-3×10^{−3})$$

```python
# Fig.8(b)
import numpy as np
import matplotlib.pyplot as plt

# Define the wavelength range and partial derivative values
lambda_1 = np.linspace(1./1000*1e9, 1./1090*1e9, 1000)
d_x = -1 * (4 * np.pi * lambda_1) * (3e-3)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(lambda_1, d_x, label=r'$∂_X\Delta k_{SHG}$')
plt.xlabel('$1/\lambda_1 (nm)$')
plt.ylabel(r'$∂_X\Delta k_{SHG}$')
plt.title(r'$∂_X\Delta k_{SHG}$ vs $1/\lambda_1$')
#plt.grid()
plt.legend()

wavelength = 1./1053*1e9
highlight_index = np.abs(lambda_1 - wavelength).argmin()
plt.scatter(lambda_1[highlight_index], d_x[highlight_index], color='black', s=100, label=f'λ_1 = {highlight_wavelength}')
plt.show()
```

Over the range of wavelengths where the partial derivatives in Eq.(a) remain relatively stable, local variations in crystal properties, such as angle and deuteration level, exhibit similar variations around phase matching. These variations are scaled by the ratio of the characterization wavelength (1053 nm) to the SHG wavelength $(\lambda_1)$. The partial derivatives in Eq.(a) undergo an approximate 10% variation over a 100-nm interval centered around 1053 nm. This suggests that characterizing the crystal properties at 1053 nm would provide a reasonable estimate of SHG performance at other wavelengths within this wavelength range.

```python
def n3(l):
  return a + b1*l**2/(l**2 - b2) + c1/(l**2 - c2)
```

## 5. Conclusions

In conclusion, the in-depth study quantifies the effect of longitudinal variations in wave-vector mismatch on second-harmonic generation and optical parametric amplification. Low-order variations exert the most significant impact on efficiency. These variations result in reduced efficiency for nonlinear processes that are nominally phase matched. Conversely, processes that are not nominally phase matched exhibit a broader range of efficiencies around the average efficiency.

The transverse variations in wave-vector mismatch measured on five DKDP crystals provide estimates of the longitudinal variations that can be expected on these crystals. Upon comparing these estimates with the simulation results, it becomes evident that the longitudinal wave-vector variations do not exert any significant influence on the process of three-wave nonlinear mixing.
